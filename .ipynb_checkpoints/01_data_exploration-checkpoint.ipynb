```python
# Feature engineering query - runs in Jupyter
feature_engineering_query = """
SELECT *,
    -- Engineer quarter from game_seconds_remaining
    CASE 
        WHEN game_seconds_remaining > 2700 THEN 1
        WHEN game_seconds_remaining > 1800 THEN 2
        WHEN game_seconds_remaining > 900 THEN 3
        ELSE 4
    END as quarter_engineered,
    
    -- Two-minute warning scenarios (your insight!)
    CASE 
        WHEN game_seconds_remaining <= 120 AND 
             (CASE 
                WHEN game_seconds_remaining > 1800 THEN 2
                WHEN game_seconds_remaining > 900 THEN 3
                ELSE 4
             END) IN (2, 4) THEN 1
        ELSE 0
    END as two_minute_scenario,
    
    -- Score differential categories
    CASE 
        WHEN ABS(score_differential) <= 3 THEN 'One Score'
        WHEN ABS(score_differential) <= 7 THEN 'Two Score'
        WHEN ABS(score_differential) <= 14 THEN 'Three Score'
        ELSE 'Blowout'
    END as game_closeness
FROM pbp_2023
"""

# Execute and save enhanced dataset
pbp_enhanced = pd.read_sql_query(feature_engineering_query, conn)
print(f"âœ… Enhanced dataset created with {len(pbp_enhanced)} records")

# Save enhanced version back to database
pbp_enhanced.to_sql('pbp_enhanced', conn, if_exists='replace', index=False)
print("âœ… Enhanced dataset saved to database")
```

```python
# Add this to your notebook - Clutch situation definitions
print("=== DEFINING CLUTCH SITUATIONS ===")

# Define multiple clutch scenarios for comprehensive analysis
clutch_definitions = {
    'fourth_quarter_close': {
        'description': '4th quarter with score difference â‰¤ 7',
        'conditions': ['quarter_engineered = 4', 'ABS(score_differential) <= 7']
    },
    'two_minute_drill': {
        'description': 'Two-minute warning scenarios (your insight!)',
        'conditions': ['two_minute_scenario = 1', 'ABS(score_differential) <= 7']
    },
    'fourth_down_conversions': {
        'description': '4th down attempts in competitive games',
        'conditions': ['down = 4', 'ABS(score_differential) <= 14']
    },
    'comeback_situations': {
        'description': 'Trailing in 4th quarter',
        'conditions': ['quarter_engineered = 4', 'score_differential < 0', 'ABS(score_differential) <= 10']
    },
    'high_leverage': {
        'description': 'High-leverage: 3rd/4th down, 4th quarter, close game',
        'conditions': ['quarter_engineered = 4', 'down IN (3,4)', 'ABS(score_differential) <= 7']
    }
}

# Print definitions for documentation
for key, definition in clutch_definitions.items():
    print(f"\n{key.upper()}:")
    print(f"  Description: {definition['description']}")
    print(f"  SQL Conditions: {' AND '.join(definition['conditions'])}")
```

```python
# Create reusable functions for clutch analysis
def create_clutch_dataset(conn, clutch_type='fourth_quarter_close'):
    """
    Create clutch dataset based on predefined situations
    """
    
    # Get conditions for this clutch type
    conditions = clutch_definitions[clutch_type]['conditions']
    where_clause = ' AND '.join(conditions)
    
    # Build comprehensive clutch query
    clutch_query = f"""
    SELECT 
        passer_player_name,
        posteam,
        game_date,
        quarter_engineered,
        down,
        ydstogo,
        game_seconds_remaining,
        score_differential,
        complete_pass,
        passing_yards,
        touchdown,
        interception,
        two_minute_scenario,
        game_closeness,
        -- Add clutch flag
        1 as is_clutch,
        '{clutch_type}' as clutch_type
    FROM pbp_enhanced
    WHERE {where_clause}
      AND passer_player_name IS NOT NULL
      AND complete_pass IS NOT NULL
      AND play_type = 'pass'
    """
    
    return pd.read_sql_query(clutch_query, conn)

def create_non_clutch_dataset(conn):
    """
    Create non-clutch dataset for comparison
    """
    non_clutch_query = """
    SELECT 
        passer_player_name,
        posteam,
        game_date,
        quarter_engineered,
        down,
        ydstogo,
        game_seconds_remaining,
        score_differential,
        complete_pass,
        passing_yards,
        touchdown,
        interception,
        two_minute_scenario,
        game_closeness,
        -- Add non-clutch flag
        0 as is_clutch,
        'regular' as clutch_type
    FROM pbp_enhanced
    WHERE quarter_engineered IN (1, 2, 3)
      AND ABS(score_differential) > 14  -- Blowout games
      AND passer_player_name IS NOT NULL
      AND complete_pass IS NOT NULL
      AND play_type = 'pass'
    """
    
    return pd.read_sql_query(non_clutch_query, conn)

# Test your functions
print("=== TESTING CLUTCH DATASET CREATION ===")
clutch_4q = create_clutch_dataset(conn, 'fourth_quarter_close')
clutch_2min = create_clutch_dataset(conn, 'two_minute_drill')
non_clutch = create_non_clutch_dataset(conn)

print(f"âœ… 4th Quarter Close: {len(clutch_4q)} plays")
print(f"âœ… Two-Minute Drill: {len(clutch_2min)} plays")
print(f"âœ… Non-Clutch: {len(non_clutch)} plays")
```

```python
# Validate your clutch definitions make sense
def validate_clutch_logic(dataset, clutch_type):
    """
    Validate that your clutch filtering worked correctly
    """
    print(f"\n=== VALIDATING {clutch_type.upper()} LOGIC ===")
    
    if len(dataset) == 0:
        print("âŒ No data found - check your filtering conditions")
        return
    
    # Check time constraints
    if 'fourth_quarter' in clutch_type:
        quarters = dataset['quarter_engineered'].unique()
        print(f"Quarters in dataset: {sorted(quarters)}")
        
    # Check score differential
    score_range = dataset['score_differential'].describe()
    print(f"Score differential range: {score_range['min']:.1f} to {score_range['max']:.1f}")
    
    # Check sample size by QB
    qb_counts = dataset['passer_player_name'].value_counts()
    print(f"QBs with most clutch attempts: {qb_counts.head(3).to_dict()}")
    
    # Check play distribution
    down_dist = dataset['down'].value_counts().sort_index()
    print(f"Down distribution: {down_dist.to_dict()}")
    
    return True

# Validate each clutch type
for clutch_type in ['fourth_quarter_close', 'two_minute_drill']:
    dataset = create_clutch_dataset(conn, clutch_type)
    validate_clutch_logic(dataset, clutch_type)

# Validate non-clutch
print("\n=== VALIDATING NON-CLUTCH LOGIC ===")
validate_clutch_logic(non_clutch, 'non_clutch')
```

```python
# Create your master datasets for analysis
print("=== CREATING MASTER DATASETS ===")

# Combine all clutch situations
all_clutch_datasets = []
for clutch_type in clutch_definitions.keys():
    dataset = create_clutch_dataset(conn, clutch_type)
    all_clutch_datasets.append(dataset)

# Master clutch dataset
master_clutch = pd.concat(all_clutch_datasets, ignore_index=True)
print(f"âœ… Master clutch dataset: {len(master_clutch)} plays")

# Save datasets for future use
master_clutch.to_sql('master_clutch', conn, if_exists='replace', index=False)
non_clutch.to_sql('master_non_clutch', conn, if_exists='replace', index=False)

# Create combined dataset with clutch flag
combined_query = """
SELECT * FROM master_clutch
UNION ALL
SELECT * FROM master_non_clutch
"""

combined_dataset = pd.read_sql_query(combined_query, conn)
combined_dataset.to_sql('combined_clutch_analysis', conn, if_exists='replace', index=False)

print(f"âœ… Combined dataset: {len(combined_dataset)} plays")
print(f"   - Clutch plays: {len(master_clutch)}")
print(f"   - Non-clutch plays: {len(non_clutch)}")
```

```python
# Updated clutch definitions with better game theory
print("=== REFINED CLUTCH DEFINITIONS ===")

clutch_definitions_v2 = {
    'fourth_quarter_close': {
        'description': '4th quarter within one score (winning by 7 or losing by 8)',
        'conditions': ['quarter_engineered = 4', 'score_differential BETWEEN -8 AND 7']
    },
    'two_minute_drill': {
        'description': 'Two-minute scenarios within striking distance',
        'conditions': ['two_minute_scenario = 1', 'score_differential BETWEEN -8 AND 7']
    },
    'fourth_down_conversions': {
        'description': '4th down attempts in competitive games',
        'conditions': ['down = 4', 'score_differential BETWEEN -14 AND 14']
    },
    'comeback_situations': {
        'description': 'Trailing by 1-2 scores in 4th quarter',
        'conditions': ['quarter_engineered = 4', 'score_differential BETWEEN -14 AND -1']
    },
    'high_leverage': {
        'description': 'High-leverage: 3rd/4th down, 4th quarter, game on the line',
        'conditions': ['quarter_engineered = 4', 'down IN (3,4)', 'score_differential BETWEEN -8 AND 7']
    },
    'protect_lead': {
        'description': 'Protecting narrow lead in 4th quarter',
        'conditions': ['quarter_engineered = 4', 'score_differential BETWEEN 1 AND 7']
    }
}

# Print updated definitions
for key, definition in clutch_definitions_v2.items():
    print(f"\n{key.upper()}:")
    print(f"  Description: {definition['description']}")
    print(f"  SQL Conditions: {' AND '.join(definition['conditions'])}")
```

```python
def create_clutch_dataset_v2(conn, clutch_type='fourth_quarter_close'):
    """
    Updated clutch dataset with refined game theory
    """
    conditions = clutch_definitions_v2[clutch_type]['conditions']
    where_clause = ' AND '.join(conditions)
    
    clutch_query = f"""
    SELECT 
        passer_player_name,
        posteam,
        game_date,
        quarter_engineered,
        down,
        ydstogo,
        game_seconds_remaining,
        score_differential,
        complete_pass,
        passing_yards,
        touchdown,
        interception,
        two_minute_scenario,
        game_closeness,
        -- Add context flags
        CASE 
            WHEN score_differential > 0 THEN 'protecting_lead'
            WHEN score_differential < 0 THEN 'trailing'
            ELSE 'tied'
        END as game_situation,
        -- Add clutch intensity score
        CASE 
            WHEN two_minute_scenario = 1 AND ABS(score_differential) <= 3 THEN 5
            WHEN quarter_engineered = 4 AND ABS(score_differential) <= 3 THEN 4
            WHEN quarter_engineered = 4 AND ABS(score_differential) <= 7 THEN 3
            ELSE 2
        END as clutch_intensity,
        1 as is_clutch,
        '{clutch_type}' as clutch_type
    FROM pbp_enhanced
    WHERE {where_clause}
      AND passer_player_name IS NOT NULL
      AND complete_pass IS NOT NULL
      AND play_type = 'pass'
    """
    
    return pd.read_sql_query(clutch_query, conn)

# Test the refined definitions
print("=== TESTING REFINED CLUTCH DEFINITIONS ===")
clutch_4q_v2 = create_clutch_dataset_v2(conn, 'fourth_quarter_close')
clutch_2min_v2 = create_clutch_dataset_v2(conn, 'two_minute_drill')
protect_lead = create_clutch_dataset_v2(conn, 'protect_lead')

print(f"âœ… 4th Quarter Close (v2): {len(clutch_4q_v2)} plays")
print(f"âœ… Two-Minute Drill (v2): {len(clutch_2min_v2)} plays")
print(f"âœ… Protect Lead: {len(protect_lead)} plays")

# Compare with original
print(f"\n=== COMPARISON WITH ORIGINAL ===")
print(f"Original 4th Quarter Close: {len(clutch_4q)} plays")
print(f"Refined 4th Quarter Close: {len(clutch_4q_v2)} plays")
print(f"Increase: {len(clutch_4q_v2) - len(clutch_4q)} plays ({((len(clutch_4q_v2) - len(clutch_4q))/len(clutch_4q)*100):.1f}% more)")
```

```python
# Simplified and corrected query
print("\n=== FIXED RODGERS VS BURROW ANALYSIS ===")

rodgers_burrow_fixed = """
SELECT 
    passer_player_name,
    COUNT(*) as clutch_attempts,
    AVG(CAST(complete_pass AS FLOAT)) as completion_rate,
    AVG(CAST(passing_yards AS FLOAT)) as avg_yards,
    SUM(CAST(touchdown AS INTEGER)) as touchdowns,
    SUM(CAST(interception AS INTEGER)) as interceptions,
    -- Game situation breakdown
    SUM(CASE WHEN score_differential > 0 THEN 1 ELSE 0 END) as protecting_lead_plays,
    SUM(CASE WHEN score_differential < 0 THEN 1 ELSE 0 END) as trailing_plays,
    SUM(CASE WHEN score_differential = 0 THEN 1 ELSE 0 END) as tied_plays
FROM pbp_enhanced
WHERE quarter_engineered = 4 
  AND score_differential BETWEEN -8 AND 7
  AND passer_player_name IS NOT NULL 
  AND complete_pass IS NOT NULL
  AND (passer_player_name LIKE '%Rodgers%' OR passer_player_name LIKE '%Burrow%')
GROUP BY passer_player_name
ORDER BY clutch_attempts DESC;
"""

try:
    rodgers_burrow_result = pd.read_sql_query(rodgers_burrow_fixed, conn)
    print(rodgers_burrow_result)
except Exception as e:
    print(f"âŒ Still getting error: {e}")
    
    # Let's check the exact player names
    name_check = """
    SELECT DISTINCT passer_player_name 
    FROM pbp_enhanced 
    WHERE passer_player_name LIKE '%Rodgers%' OR passer_player_name LIKE '%Burrow%'
    """
    names = pd.read_sql_query(name_check, conn)
    print("\nActual player names in dataset:")
    print(names)
```

```python
# If SQL is giving issues, let's do this in pandas (often easier for debugging)
print("\n=== ALTERNATIVE PANDAS APPROACH ===")

# Filter for clutch situations in pandas
clutch_plays = pbp_enhanced[
    (pbp_enhanced['quarter_engineered'] == 4) & 
    (pbp_enhanced['score_differential'].between(-8, 7)) &
    (pbp_enhanced['passer_player_name'].notna()) &
    (pbp_enhanced['complete_pass'].notna())
].copy()

print(f"Total clutch plays: {len(clutch_plays)}")

# Check for Rodgers and Burrow
rodgers_burrow_names = clutch_plays['passer_player_name'].unique()
target_qbs = [name for name in rodgers_burrow_names if 'Rodgers' in name or 'Burrow' in name]
print(f"Found QBs: {target_qbs}")

if target_qbs:
    # Analyze Rodgers and Burrow
    rb_analysis = clutch_plays[clutch_plays['passer_player_name'].isin(target_qbs)].groupby('passer_player_name').agg({
        'complete_pass': ['count', 'mean'],
        'passing_yards': 'mean',
        'touchdown': 'sum',
        'interception': 'sum',
        'score_differential': lambda x: (x > 0).sum()  # protecting lead plays
    }).round(3)
    
    print("\n=== RODGERS vs BURROW CLUTCH STATS ===")
    print(rb_analysis)
```
